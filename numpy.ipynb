Trabalho 3 - Numpy

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Salary_Data.csv')
print(df.head())
```

```python
years_of_experience = df['YearsExperience'].to_numpy()
salaries = df['Salary'].to_numpy()
print(years_of_experience)
print(salaries)
```

```python
mean_salary = np.mean(salaries)
print(f'Média dos salários: {mean_salary}')
```

```python
mean_years_experience = np.mean(years_of_experience)
print(f'Média de tempo de serviço: {mean_years_experience}')
```

```python
salaries_above_5_years = salaries[years_of_experience > 5]
mean_salary_above_5_years = np.mean(salaries_above_5_years)
print(f'Salário médio dos funcionários com mais de 5 anos de serviço: {mean_salary_above_5_years}')
```

```python
plt.figure(figsize=(10,6))
plt.bar(years_of_experience, salaries, color='blue')
plt.title('Salários vs. Anos de Experiência')
plt.xlabel('Anos de Experiência')
plt.ylabel('Salário')
plt.show()
```

```python
increased_salaries = salaries * 1.10
df['IncreasedSalary'] = increased_salaries
print(f'Salários após aumento de 10%:\n{df[["YearsExperience", "Salary", "IncreasedSalary"]]}')
```
